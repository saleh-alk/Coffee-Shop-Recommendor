```python
######Content Based Recommendation#######

from sklearn.feature_extraction.text import TfidfVectorizer

# Using Abhishek Thakur's arguments for TF-IDF
tfv = TfidfVectorizer(min_df=3,  max_features=None, 
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')
tfv_matrix = tfv.fit_transform(df['reviews'])
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel


sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
indices = pd.Series(df.index, index=df['shop_name']).drop_duplicates()
```

```python
def give_rec(title, sig=sig):
    # Get the index corresponding to original_title
    idx = indices[title]

    
    ## got the pairwise simularity score
    sig_scores = list(enumerate(sig[idx]))

   ## sorted the coffee shop_name
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    ## got the first ten
    sig_scores = sig_scores[1:11]

    
    final_indices = [i[0] for i in sig_scores]

    
    return df['shop_name'].iloc[final_indices]
```

```python
give_rec('The Shaky')
```

```python
### Nearest Neighbor filtering
```

```python
df = df.reset_index()

df['New_ID'] = df.index 
```

```python
popularity_threshold = 100
rating_popular_shops = df.query('numbers_of_rating >= @popularity_threshold')
rating_popular_shops.head()
```

```python
shop_features_df=rating_popular_shops.pivot_table(index='shop_name', columns = 'index' , values='rating').fillna(0)
shop_features_df.head()
```

```python
from scipy.sparse import csr_matrix

df_matrix = csr_matrix(shop_features_df.values)

from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(df_matrix)
```

```python
query_index = np.random.choice(shop_features_df.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(shop_features_df.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(shop_features_df.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, shop_features_df.index[indices.flatten()[i]], distances.flatten()[i]))
```
